```python
def read(
    channel = 0, 
    device = 0,
    ):

    # voltage array
    vou = np.empty(10000)

    for i in range(50000):
        try:
            vou[i] = adc.readADC(channel=channel, device=device)

        except:
            vou.resize(vou.shape[0] * 2)

    return vou 

@np.vectorize
def resistance_from_voltage(
    R2,
    Vin,
    Vou
    ):
    return np.multiply(R2, (np.subtract(np.divide(Vin, Vou), 1)))
```

```python
vou = read(channel = 0, device = 0)

resistance = resistance_from_voltage(Vou = vou, Vin = 3.3, R2 = 10.0)

pressure_df = pd.DataFrame(data=[resistance, vou], index=['resistance', 'voltage']).T

pressure_df.replace([np.inf, -np.inf], 0, inplace=True)
```

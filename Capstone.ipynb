```python
from accelerometers import Accelerometers
from resistanceSensors import ResistanceSensors
import time


accs = Accelerometers()
pressureSensors = ResistanceSensors(0)

tref = time.time()
pressureSensors.calibrate(tref)
accs.calibrate(tref)

while time.time() - tref < 60:
    accs.record()
    pressureSensors.record()


pressureSensors.save("pressureSensors4.npy")
accs.save("accelerometers4.npy")

accs.table().head()
```

```python
from accelerometers import Accelerometers
from resistanceSensors import ResistanceSensors
import time


accs = Accelerometers()
flexSensors = ResistanceSensors(0)

tref = time.time()
flexSensors.calibrate(tref)
accs.calibrate(tref)

while time.time() - tref < 60:
    accs.record()
    # flexSensors.record()
    pressureSensors.record()

flexSensors.save("pressureSensors.npy")
accs.save("accelerometers2.npy")
```
